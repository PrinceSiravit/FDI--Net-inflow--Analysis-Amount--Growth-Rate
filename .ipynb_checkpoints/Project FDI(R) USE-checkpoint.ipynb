###  Finding average growth rate of each country.

Finding the annual growth rate of each country

Since pct_change is used, the data must be in float format, so string columns must be dropped.

```python
dfyoy = dfuse.drop(['Country Name','IncomeGroup'], axis=1, inplace=False)
dfchange = dfyoy.pct_change(axis=1)

dfchange['Average'] = dfchange.mean(axis=1, numeric_only = True)
```

```python
dfchange
```

### Insert Column

Add an Income group column to use for sorting.

```python
extract = dfuse['IncomeGroup']

dfchange = dfchange.join(extract)
```

```python
dfchange
```

### Organization

Organize information for easy reading and understanding.

```python
first_column = dfchange.pop('IncomeGroup') 

dfchange.insert(0,'IncomeGroup',first_column)
```

```python
dfchange
```

### Finding Average FDI growth rate of Each group

```python
dfchange.groupby(['IncomeGroup']).mean()
```

```python
dfgraph1 = dfchange.groupby(['IncomeGroup']).mean()
```

### Finding Average of FDI of each country

```python
dfuse['Average'] = dfuse.mean(axis=1, numeric_only = True)
```

```python
dfuse
```

### Finding Average of FDI of each group

```python
#Coding...
dfuse.groupby(['IncomeGroup']).mean()
```

```python
dfgraph2 = dfuse.groupby(['IncomeGroup']).mean()
```

### Key Findings

#### Figure 1: Averge YoY of each Income Group

```python
import matplotlib as mpl
```

```python
dfgraph1['Average'].T.plot(kind='bar')
```

#### Figure1

It shows the average growth Foreign Direct Investment(FDI) of each group can represent economic growth, inflation, exchange rates, and population growth of each group.

The average growth of each group:
High Income Group = -1.41%
Low Income Group = 2.5%
Lower Middle Income Group = 1.65%
Upper Middle Income Group = 0.81%

Key Finding: The group with the highest average growth rate was the Lower Income group, followed by Lower Middle Income and Upper Middle Income respectively, but the group with the greatest decline in average growth was High Income.

#### Figure 2: Average FDI proportion of each income group

```python
dfgraph2['Average'].plot.pie(autopct="%1.1f%%");
```

#### Figure 2

It shows the proportion of Foreign Direct Investment(FDI) of each group can represent the proportion of concentration of prosperity.

High Income Group = 67.4%
Low Income Group = 1.2%
Lower Middle Income Group = 5.5%
Upper Middle Income Group = 26.0%

Key Finding: 
The proportion of FDI shows the ratio of investments concentrated in the high income and Upper middle income segments, demonstrating the investment trend of foreign investors and the growth of countries in these two groups.

#### Figure 3: FDI trend of each Income Group

```python
dfgraph2_2 = dfgraph2.drop('Average', axis=1, inplace=False)

dfgraph2_2.T.plot()
```

#### Figure 3

It shows the trend of Foreign Direct Investment(FDI) of each group can represent the trend of economic growth, inflation, exchange rates, and population growth of each group.

Key Finding: The trend of FDI shows all of the line charts are oriented in the same trend, differing in volatility. It can be noted that the volatility of High Income is the highest followed by  Upper Middle Income, Lower Middle Income, and Low Income.

## Significance

The relationship between average amount of Foreign Direct Investment and average Growth rate of income group has a significant relationship.

When the results of figure 1 and figure 2 were matched together, it shows:

Growth Rate:
High Income Group = -1.41%
Low Income Group = 2.5%
Lower Middle Income Group = 1.65%
Upper Middle Income Group = 0.81%

FDI proportion:
High Income Group = 67.4%
Low Income Group = 1.2%
Lower Middle Income Group = 5.5%
Upper Middle Income Group = 26.0%

From the match, it indicates that groups with high FDI will have regressive growth rates. But those with low FDI had high growth rates. Thus, the amount of FDI and the growth rate of FDI of each Income group has a oppsite relationship.
