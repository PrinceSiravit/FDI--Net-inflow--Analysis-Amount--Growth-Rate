###  Finding average growth rate of each country.

Finding the annual growth rate of each country

Since pct_change is used, the data must be in float format, so string columns must be dropped.

```python
dfyoy = dfuse.drop(['Country Name','IncomeGroup'], axis=1, inplace=False)
dfchange = dfyoy.pct_change(axis=1)

dfchange['Average'] = dfchange.mean(axis=1, numeric_only = True)
```

```python
dfchange
```

### Insert Column

Add an Income group column to use for sorting.

```python
extract = dfuse['IncomeGroup']

dfchange = dfchange.join(extract)
```

```python
dfchange
```

### Organization

Organize information for easy reading and understanding.

```python
first_column = dfchange.pop('IncomeGroup') 

dfchange.insert(0,'IncomeGroup',first_column)
```
